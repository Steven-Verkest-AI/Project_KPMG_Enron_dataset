# Sent allen-p

```python
sent = pd.DataFrame()
sent["File"] = listdir(path + "sent")
```

```python
def read_content(directory, file):
    with open(join(path, directory, file), "r") as text:
        return text.read()
sent["Raw"] = sent["File"].apply(lambda x: read_content("sent", x))
```

```python
email.message_from_string(sent.loc[0, "Raw"]).get("X-Folder")
```

```python
sent["From"] = sent["Raw"].apply(lambda x: email.message_from_string(x).get("X-From"))
sent["To"] = sent["Raw"].apply(lambda x: email.message_from_string(x).get("X-To"))
sent["Subject"] = sent["Raw"].apply(lambda x: email.message_from_string(x).get("Subject"))
sent["Date"] = sent["Raw"].apply(lambda x: email.message_from_string(x).get("Date"))
sent["Body"] = sent["Raw"].apply(lambda x: email.message_from_string(x).get_payload())
```
